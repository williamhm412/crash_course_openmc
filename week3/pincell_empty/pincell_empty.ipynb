# Create a Pincell
Your task is to use the elements of OpenMC that we've covered in the last 2 weeks and create a simple pincell.

### The specifications of the pincell are as follows:
- The pincell will be infinite in the z direction, and a box in the xy direction
- Center the pincell at (0,0) in the xy plane
- Inside the box will be a fuel cylinder nested inside a thin gap cylinder which is nested within a cladding cylinder
    - The radius of the fuel cylinder will be 0.39 cm
    - The outer radius of the gap cylinder will be 0.4 cm
    - The outer radius of the clad cylinder will be 0.46 cm
    - The box will have side lengths of 1.26
- The boundaries of the box will be set to reflect neutrons
- run with the following settings
    - a point source at (0,0,0)
    - 1000 particles per cycle
    - 100 active cycles
    - 20 inactive cycles
- The following materials will be used
    - Water (the moderator) with only H1 and O16 at 1 g/cm^3 density and 293 K
    - Pure elemental Zirconium (the clad) at 6.6 g/cm^3 density and 293 K
    - UO2 (the fuel) with 2 atoms Oxygen, to 1 atom Uranium with the Uranium being 3% U235 and 97% U238 all at 10 g/cm^3 and 293 K
